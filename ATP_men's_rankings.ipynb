```python
import json
import requests
import pandas as pd
```

```python
url = 'https://api.sofascore.com/api/v1/rankings/type/5'
```

```python
req = requests.get(url)
json_raw = json.loads(req.content)
```

```python
json_raw
```

```python
json_raw.keys()
```

```python
json_raw["rankings"]
```

```python
data = pd.json_normalize(json_raw['rankings'])
```
